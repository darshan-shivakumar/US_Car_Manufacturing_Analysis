# US Car Manufacturing analysis

```python
# Let's import the modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
car_data= pd.read_csv('Car dataset.csv')
car_data.head()
```

```python
# Let's remove the unwanted columns

car_data.drop(['Unnamed: 0','vin','lot','condition'],axis=1,inplace=True)
car_data.head()
```

## 1. Top 5 Car brands in terms of number of cars

```python
brands=[]

for i in car_data['brand'].unique():
    brands.append([i, len(car_data[car_data['brand']== i])])
br_data= pd.DataFrame(brands, columns=['Brand','Frequency']).sort_values(by='Frequency', ascending= False) 
br_data.head()
```

```python
br_data.head().plot(x='Brand', y='Frequency',kind='bar',figsize=(5,5))
```

### We can conclude that Ford, Dodge, Nissan, Chevrolet, Gmc are top 5 car brands in terms of number of cars.

## 2. Visualisation on Car brand based on frequency of cars

```python
cars=[]

for i in car_data['brand'].unique():
    cars.append([i,len(car_data[car_data['brand']==i])])

cars_data= pd.DataFrame(cars,columns=['Brands','Frequency']).sort_values(by='Frequency',ascending=False)
cars_data
```

```python
cars_data.plot(x='Brands',y='Frequency',kind='bar',figsize=(20,10),color='purple')
```

### Here's a Visualisation on Car brand based on frequency of cars.

## 3. Visualize  Car Industry Trend

```python
year=[]

for i in car_data['year'].unique():
    year.append([i,len(car_data[car_data['year']==i])])

trend= pd.DataFrame(year,columns=['year','frequency']).sort_values(by='year',ascending=True)
trend
```

```python
trend.plot(x='year',y='frequency',kind='line',figsize=(5,5),color='darkblue')
```

### Here we can figure out that from year 2010 there's been increase in car sales every year upto 2019 and then a decrease in the year 2020.

## 4. Most Popular Car Colours 

```python
color=[]

for i in car_data['color'].unique():
    color.append([i,len(car_data[car_data['color']==i])])
pop_color= pd.DataFrame(color,columns=['color','frequency']).sort_values(by='frequency',ascending=False)
pop_color
```

```python
pop_color.plot(x='color',y='frequency',kind='bar',figsize=(20,10),color='violet')
```

### We can conclude that white, black, grey, silver, red, blue etc. are some of the most popular car colours

## 5. Which car brand is having Maximum Average Price 

```python
# here we have multiple columns i.e brand and price so we use groupby() function

car_price= car_data.groupby('brand').mean()['price'].sort_values(ascending=False)
car_price
```

```python
car_price.plot.bar(figsize=(20,10),color='orange')
```

###  We can conclude that Harley-Davidson brand is having Maximum Average Price

## 6. Which State is having most expensive cars 

```python
# here we have two columns state and cars so let's use groupby() function

exp_car=car_data.groupby('state').mean()['price'].sort_values(ascending = False)
exp_car
```

```python
exp_car.plot.bar(figsize=(20,10),color='red')
```

### We can conclude that the state 'Kentucky' is having the most expensive cars.

## 7.  Which Year is having Maximum Average Price

```python
year= car_data.groupby('year').mean()['price'].sort_values(ascending=False)
year
```

```python
year.plot.bar(figsize=(20,10),color= 'brown')
```

### We can conclude that the year 1973 is having Maximum Average Price.
